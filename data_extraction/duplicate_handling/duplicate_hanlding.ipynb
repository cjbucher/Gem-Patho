```python
gene_list_path = "/home/chb3333/yulab/chb3333/gem-patho/data_extraction/cancer_gene_list_selection/combined_genelist.csv"
gene_df = pd.read_csv(gene_list_path)
# Extract the unique gene symbols as a set for fast lookup
gene_set = set(gene_df["Gene Symbol"].dropna().unique())
print(f"Loaded {len(gene_set)} genes from the gene list.")
```

```python
# Set the base directory for the WXS folders containing raw mutation files.
BASE_DIR = "/home/chb3333/scratch/GDC-WGS-WXS/wxs"

def get_sample_metrics(file_id, file_name):
    """
    For a given sample (defined by its folder 'file_id' and raw mutation file 'file_name'),
    returns a dictionary with:
      - flagged: True if an annotations.txt file exists in the folder and any row's entity_type contains "masked_somatic_mutation".
      - mutation_count: Count of mutation rows in the raw mutation file (skipping comment lines)
                        where the Hugo_Symbol (from the "Hugo_Symbol" column) matches one in gene_set.
      - created_datetime: The maximum created_datetime found in the annotations.txt file (if available).
    
    If the raw mutation file is gzipped (i.e. file_name ends with '.gz'), it is opened with gzip.
    """
    sample_info = {"flagged": False, "mutation_count": 0, "created_datetime": None}
    
    # --- Process annotations.txt ---
    ann_path = os.path.join(BASE_DIR, file_id, "annotations.txt")
    if os.path.exists(ann_path):
        try:
            ann_df = pd.read_csv(ann_path, sep="\t")
            if "entity_type" in ann_df.columns:
                masked_rows = ann_df[ann_df["entity_type"].str.contains("masked_somatic_mutation", na=False)]
                if not masked_rows.empty:
                    sample_info["flagged"] = True
                if "created_datetime" in ann_df.columns:
                    ann_df["created_datetime_parsed"] = pd.to_datetime(ann_df["created_datetime"], errors="coerce")
                    sample_info["created_datetime"] = ann_df["created_datetime_parsed"].max()
            else:
                if "created_datetime" in ann_df.columns:
                    ann_df["created_datetime_parsed"] = pd.to_datetime(ann_df["created_datetime"], errors="coerce")
                    sample_info["created_datetime"] = ann_df["created_datetime_parsed"].max()
        except Exception as e:
            print(f"Error reading annotations.txt for {file_id}: {e}")
    
    # --- Count mutations from the raw mutation file ---
    mutation_file = os.path.join(BASE_DIR, file_id, file_name)
    if os.path.exists(mutation_file):
        try:
            # If the file is gzipped, use gzip.open in text mode.
            if file_name.endswith(".gz"):
                f_open = lambda f: gzip.open(f, "rt", encoding="utf-8")
            else:
                f_open = lambda f: open(f, "r", encoding="utf-8")
                
            with f_open(mutation_file) as f:
                header = None
                hugo_idx = None
                count = 0
                for line in f:
                    line = line.strip()
                    # Skip comment lines starting with "#"
                    if line.startswith("#"):
                        continue
                    # First non-comment line is the header.
                    if header is None:
                        header = line.split("\t")
                        try:
                            hugo_idx = header.index("Hugo_Symbol")
                        except ValueError:
                            print(f"'Hugo_Symbol' column not found in header for file {mutation_file}")
                            hugo_idx = None
                        continue
                    # Process mutation rows.
                    if header is not None and hugo_idx is not None:
                        fields = line.split("\t")
                        if len(fields) > hugo_idx:
                            hugo_symbol = fields[hugo_idx]
                            if hugo_symbol in gene_set:
                                count += 1
            sample_info["mutation_count"] = count
        except Exception as e:
            print(f"Error reading mutation file for {file_id}: {e}")
    else:
        print(f"Mutation file {mutation_file} does not exist for {file_id}")
    
    return sample_info
```

```python
# Cell 2: Compute metrics for each unique File ID in duplicate_wxs and add new columns.
unique_pairs = duplicate_wxs[["File ID", "File Name"]].drop_duplicates()

metrics_dict = {}
for idx, row in unique_pairs.iterrows():
    fid = row["File ID"]
    fname = row["File Name"]
    metrics_dict[(fid, fname)] = get_sample_metrics(fid, fname)

# Now update the duplicate_wxs DataFrame using .loc to avoid SettingWithCopyWarning.
duplicate_wxs.loc[:, "flagged"] = duplicate_wxs.apply(
    lambda r: metrics_dict.get((r["File ID"], r["File Name"]), {}).get("flagged", False), axis=1)
duplicate_wxs.loc[:, "mutation_count"] = duplicate_wxs.apply(
    lambda r: metrics_dict.get((r["File ID"], r["File Name"]), {}).get("mutation_count", 0), axis=1)
duplicate_wxs.loc[:, "created_datetime"] = duplicate_wxs.apply(
    lambda r: metrics_dict.get((r["File ID"], r["File Name"]), {}).get("created_datetime", pd.NaT), axis=1)

# Optional: display the first few rows to verify the new columns
display(duplicate_wxs.head())
```

```python
duplicate_wxs.loc[:, "created_datetime_naive"] = duplicate_wxs["created_datetime"].apply(
    lambda x: x.tz_localize(None) if pd.notnull(x) and isinstance(x, pd.Timestamp) and x.tz is not None else x
)

def sort_samples(df):
    # Sort by flagged (True first), then by mutation_count (descending),
    # and then by created_datetime_naive (descending)
    df_sorted = df.sort_values(
        by=["flagged", "mutation_count", "created_datetime_naive"],
        ascending=[False, False, False]
    )
    return df_sorted

# For each Case ID, select the sample with the highest priority.
priority_df = duplicate_wxs.groupby("Case ID").apply(lambda group: sort_samples(group).iloc[0]).reset_index(drop=True)

print("Recommended sample (File ID) per Case ID:")
display(priority_df)
```

```python
# Deduplicate wxs_sample_sheet

sample_sheet_path = "/home/chb3333/yulab/chb3333/gem-patho/data_extraction/wxs_sample_sheet_clean.tsv"
output_path = "/home/chb3333/yulab/chb3333/gem-patho/data_extraction/deduplicated_wxs_sample_sheet_clean.tsv"
sample_sheet = pd.read_csv(sample_sheet_path, sep="\t")
print(f"Loaded sample sheet with {sample_sheet.shape[0]} rows.")
```

```python
priority_csv = "/home/chb3333/yulab/chb3333/gem-patho/data_extraction/duplicate_handling/highest_priority_samples.csv"
priority_df = pd.read_csv(priority_csv)
print(f"Loaded highest priority selection for {priority_df.shape[0]} duplicate Case IDs.")
```

```python
# Merge the sample sheet with the priority_df based on "Case ID".
# For cases with duplicates, priority_df has the recommended "File ID".
merged = sample_sheet.merge(priority_df, on="Case ID", how="left", suffixes=("", "_priority"))

# For rows where "File ID_priority" is not null, we know that the Case ID is duplicated.
# We keep the row only if its "File ID" matches the recommended one.
dedup = merged[(merged["File ID_priority"].isnull()) | (merged["File ID"] == merged["File ID_priority"])].copy()
```

```python
dedup.drop(columns=["File ID_priority"], inplace=True)

print(f"Deduplicated sample sheet has {dedup.shape[0]} rows.")
```

```python
dedup = dedup[['File ID', 'File Name', 'Data Category', 'Data Type', 'Project ID', 'Case ID', 'Sample ID', 'Sample Type']]
```

```python
dedup.to_csv(output_path, sep="\t", index=False)
print(f"Saved deduplicated sample sheet to: {output_path}")
```
